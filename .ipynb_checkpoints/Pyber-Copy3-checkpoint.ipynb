```python
# Dependencies
import pandas as pd
import os
import numpy as np
import io
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load in City and Ride data from csv files
city_path = os.path.join('raw_data', 'city_data.csv')
city = pd.read_csv(city_path)
city.head()
```

```python
ride_path = os.path.join('raw_data', 'ride_data.csv')
ride = pd.read_csv(ride_path)
ride.head()
```

```python
city_data = pd.merge(city, ride, how='inner', on='city')
city_data = city_data.sort_values(['type'])
city_data = city_data[['city','date','fare','ride_id','driver_count', 'type']]
city_data = city_data.reset_index()
type_color = city_data['type']
 
# Replace --> Rural: Gold, Suburban: Light Sky Blue, and Urban: Light Coral
# type_color=type_color.replace({'Rural': 'Gold','Suburban': 'Light Sky Blue','Urban': 'Light Coral'}) 
type_color=type_color.replace({'Rural': 1,'Suburban': 3,'Urban': 5})
city_data['Type_Color'] = type_color
city_data = city_data.drop(['index'],axis = 1)
city_data.head()
```

```python
group_by_city = city_data.groupby('city')
group_by_city.count()
```

```python
avg_fare_by_city = group_by_city["fare"].mean()
avg_fare_by_city.head()
# type(avg_fare_by_city)
y_axis = avg_fare_by_city.values
y_axis[0]
```

```python
rides_per_city = group_by_city["ride_id"].count()
x_axis = rides_per_city.values
```

```python
drivers_by_city = group_by_city["driver_count"].mean()
circle_sizes = drivers_by_city.values 
circle_sizes[0]
```

```python
type_by_city = group_by_city["Type_Color"].mean()
type_by_city = type_by_city.replace({1: 'Gold', 3: 'Light Sky Blue', 5: 'Light Coral'}) 
color_names = type_by_city.values
```

```python
plot_values = pd.DataFrame({
                 "x":rides_per_city,
                 "y":avg_fare_by_city,
                 "s":drivers_by_city,
                 "t":type_by_city
                 })
plot_values[plot_values['t'] == 'Light Coral']
```

```python
type_by_city = group_by_city["Type_Color"].mean()
type_by_city = color_by_city.replace({1: 'Gold', 3: 'Light Sky Blue', 5: 'Light Coral'}) 
color_names = type_by_city.values
color_names[0]
```

```python
plt.scatter(x_axis, y_axis, s=circle_sizes*5, c = 'Gold', edgecolors="grey", linewidth=2)
plt.show()
```
